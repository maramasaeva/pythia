```python
# moral stories files preprocessing
# these files are in jsonl format, need to convert them to csv's first (doesnt work with code so just doing it with a converter)
# just these folders, like mentioned in delphi paper: 
# classification/action+context, classification/action+context+consequence, generation/action_context, generation/action_context+consequence
```

```python
import pandas as pd
```

```python
df = pd.read_csv('moral_stories_full.csv', header=0)
df
```

```python
# delete columns we dont need: ID, norm (could be useful but delphi didnt use it), moral consequence, immoral consequence
```

```python
df = df[['situation','intention','moral_action', 'immoral_action']]
df
```

```python
# we want to concatenate these two columns, but alternating the rows 
new_df = df[['moral_action','immoral_action']]
new_df
```

```python
# first action of pair is moral, second is immoral
new_df = pd.DataFrame(data=new_df.values.reshape(df.shape[0]*2,-1), columns=['action_pairs'])
```

```python
new_df
```

```python
import numpy as np
# we want to take the situation and intention column and double every row (so each row appears twice, paired under each other)
situationcolumn = df[["situation"]]
situationcolumn
```

```python
# replicating the columns
s_df = pd.DataFrame(np.repeat(situationcolumn.values, 2, axis=0))
s_df
```

```python
# column name changed to int 0, changing that now
s_df.rename(columns = {0 : 'situation'}, inplace = True)
s_df
```

```python
# adding these columns back to the original df creates problems with NaN's, so we'll just create a new df with the new columns
# (which are double the length of the old columns)
```

```python
# now do the same for intention column
intentioncolumn = df["intention"]
i_df = pd.DataFrame(np.repeat(intentioncolumn.values, 2, axis=0))
```

```python
# add this column to the situationcolumn df
s_df = s_df.join(i_df)
s_df
```

```python
# renaming intention column
s_df.rename(columns = {0 : 'intention'}, inplace = True)
s_df
```

```python
new_df
```

```python
s_df
```

```python
# add actionpairs to df
s_df = s_df.join(new_df)
s_df
```

```python
# creating new column with 'good' in every row, but we need to change every second row to 'bad'
s_df["judgement"]= 'good'
```

```python
s_df
```

```python
# changing every second row to bad here, and it works!
s_df['judgement'] = np.where(s_df.index % 2, 'bad', 'good')
```

```python
s_df
```

```python
# last step is to create train, dev and test sets
```

```python
df_trn, df_dev, df_tst = np.split(s_df.sample(frac=1, random_state=42), [int(.6*len(s_df)), int(.8*len(s_df))])
```

```python
moralstories_freeform_train = df_trn
```

```python
moralstories_freeform_dev = df_dev
```

```python
moralstories_freeform_test = df_tst
```
