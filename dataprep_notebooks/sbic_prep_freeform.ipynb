```python
# maybe i should do it differently: change intentYN and offensive YN values to categories
# (>= 0.5 means its bad, < 0.5 means its okay)
# and keep the rows where the values of those two columns are the same. so youre taking both columns into account!

# df_train_v1.loc[(df_train_v1.intentYN >= 0.5),'intentYN']='offensive'
# df_train_v1.loc[(df_train_v1.intentYN < 0.5),'intentYN']="it's okay"

df_train_v1.loc[df_train_v1['intentYN'] >= 0.5, 'intentYN'] = 1
```

```python
df_train_v1.loc[df_train_v1['intentYN'] < 0.5, 'intentYN'] = 0
```

```python
df_train_v1.loc[df_train_v1['offensiveYN'] >= 0.5, 'offensiveYN'] = 1
df_train_v1.loc[df_train_v1['offensiveYN'] < 0.5, 'offensiveYN'] = 0
```

```python
df_train_v1.dtypes
```

```python
df_train_v1.isna().sum()
```

```python
df_train_v1 = df_train_v1.astype({"intentYN":'int', "offensiveYN":'int'}) 
```

```python
df_train_v1.dtypes
```

```python
df_train_v1.intentYN.replace((1, 0), ("it's offensive", "it's okay"), inplace=True)
```

```python
df_train_v1.offensiveYN.replace((1, 0), ("it's offensive", "it's okay"), inplace=True)
```

```python
df_train_v1 = df_train_v1[df_train_v1['intentYN'] == df_train_v1['offensiveYN']]
```

```python
# just delete one of the two judgement columns now, since they both say the same
# also change name, this is final version
df_train = df_train_v1.drop('offensiveYN', 1)
```

```python
# now do the same for the dev and test sets
df_dev_v1.dtypes

df_dev_v1.loc[df_dev_v1['intentYN'] >= 0.5, 'intentYN'] = 1
df_dev_v1.loc[df_dev_v1['intentYN'] < 0.5, 'intentYN'] = 0

df_dev_v1.loc[df_dev_v1['offensiveYN'] >= 0.5, 'offensiveYN'] = 1
df_dev_v1.loc[df_dev_v1['offensiveYN'] < 0.5, 'offensiveYN'] = 0
```

```python
df_dev_v1.intentYN.replace((1, 0), ("it's offensive", "it's okay"), inplace=True)
df_dev_v1.offensiveYN.replace((1, 0), ("it's offensive", "it's okay"), inplace=True)

df_dev_v1 = df_dev_v1[df_dev_v1['intentYN'] == df_dev_v1['offensiveYN']]
```

```python
df_dev = df_dev_v1.drop('offensiveYN', 1)
df_dev
```

```python
df_test_v1.dtypes

df_test_v1.loc[df_test_v1['intentYN'] >= 0.5, 'intentYN'] = 1
df_test_v1.loc[df_test_v1['intentYN'] < 0.5, 'intentYN'] = 0

df_test_v1.loc[df_test_v1['offensiveYN'] >= 0.5, 'offensiveYN'] = 1
df_test_v1.loc[df_test_v1['offensiveYN'] < 0.5, 'offensiveYN'] = 0
```
