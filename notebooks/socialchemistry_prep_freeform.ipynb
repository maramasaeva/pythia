```python
# create 9 dataframes total: 1 for each task, and 1 for each split in that task

import pandas as pd
```

```python
df_nosplit = pd.read_csv('social-chem-101.v1.0.tsv', sep='\t', header=0)
```

```python
# split dataset in three: train, test and dev set. 

df_nosplit.sample(n=3)
```

```python
list(df_nosplit)
```

```python
# delete the columns we don't need. we need: situation (scraped from web), action (both in examples), rot-judgement (about action)

df_nosplit = df_nosplit[['split','situation','action', 'rot-judgment']]
```

```python
df_train = df_nosplit[df_nosplit['split'] == 'train']
df_train.sample(n=5)
```

```python
df_dev = df_nosplit[df_nosplit['split'] == 'dev']
df_test = df_nosplit[df_nosplit['split'] == 'test']

df_dev.sample(n=5)
```

```python
# situation and action should be make into one column, since they say the same thing but in other words. their judgements shouldnt change!

selected_columns = df_train[["action","rot-judgment"]]
df_train_2 = selected_columns.copy()

df_train_2
```

```python
df_train
```

```python
df_train.drop('split', inplace=True, axis=1)
```

```python
df_train
```

```python
df_train.drop('action', inplace=True, axis=1)
```

```python
df_train
```

```python
df_train_2 = df_train_2.rename(columns={'action': 'situation'})
```

```python
df_train_2
```

```python
df_train = pd.concat([df_train,df_train_2], axis=0, ignore_index=True)
```

```python
df_train

# this works but could be done more efficiently... dont know how tho. maybe you should make this into a function?
```

```python
# doing the same for dev and test set

selected_columns = df_dev[["action","rot-judgment"]]
df_dev_2 = selected_columns.copy()

df_dev_2
```

```python
df_dev.drop('split', inplace=True, axis=1)
df_dev.drop('action', inplace=True, axis=1)

df_dev
```

```python
df_dev_2 = df_dev_2.rename(columns={'action': 'situation'})
```

```python
df_dev = pd.concat([df_dev,df_dev_2], axis=0, ignore_index=True)
```

```python
selected_columns = df_test[["action","rot-judgment"]]
df_test_2 = selected_columns.copy()

df_test_2
```

```python
df_test.drop('split', inplace=True, axis=1)
df_test.drop('action', inplace=True, axis=1)

df_test
```

```python
df_test_2 = df_test_2.rename(columns={'action': 'situation'})
```

```python
df_test = pd.concat([df_test,df_test_2], axis=0, ignore_index=True)
df_test
```

```python
# we now have three df's consisting of a situation and a judgment about that situation
```

```python
socialchemistry_freeform_train = df_train
```

```python
socialchemistry_freeform_dev = df_dev
```

```python
socialchemistry_freeform_test = df_test
```
